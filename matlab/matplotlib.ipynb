```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,5,8,9,10]

#line plot
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Basic line plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,5,8,9,10]

#customize line plot
plt.plot(x,y,color='red',linestyle='-',marker='4',linewidth="1",markersize=6)
plt.grid(True)
```

```python
#muliple plots
##sample Data
x=[1,2,3,4,5]
y1=[1,4,7,9,15]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x,y2,color='yellow')
plt.title("Plot 4")
```

```python
#bar plot
categories=['A','B','C','D','E']
values=[5,7,3,8,6]

#create a bar plot
plt.bar(categories,values,color='blue')
```

```python
data = [1,2,3,4,4,5,6,6,6,7,7,7]

plt.hist(
    data,
    bins=5,                  # number of bins (or sequence of bin edges)
    color='orange',          # bar fill color
    edgecolor='black',       # border color
    alpha=0.7,               # transparency (0 to 1)
    histtype='bar',          # 'bar', 'step', 'stepfilled'
    orientation='vertical',  # 'vertical' or 'horizontal'
    rwidth=0.9,              # relative bar width (0 to 1)
    density=False,           # if True, normalizes histogram
    cumulative=False,        # if True, cumulative distribution
    label="My Data"          # legend label
)

# Add titles & labels
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

plt.scatter(
    x, y,
    color='blue',        # point color
    marker='x',          # shape of points
    s=100,               # size of points
    alpha=0.7,           # transparency (0 to 1)
    edgecolors='black',  # border color around markers
    linewidths=1.5,      # thickness of marker edge
    label="Data Points"  # for legend
)

# Labels & title
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellow', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  # separate A slice

plt.pie(
    sizes,
    labels=labels,        # names of slices
    colors=colors,        # custom colors
    autopct='%1.1f%%',    # show % inside slices
    counterclock=True, shadow=True   # slices direction
)

plt.title("Pie Chart Example")
plt.show()
```
