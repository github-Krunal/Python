```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,5,8,9,10]

#line plot
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Basic line plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,5,8,9,10]

#customize line plot
plt.plot(x,y,color='red',linestyle='-',marker='4',linewidth="1",markersize=6)
plt.grid(True)
```

```python
#muliple plots
##sample Data
x=[1,2,3,4,5]
y1=[1,4,7,9,15]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x,y2,color='yellow')
plt.title("Plot 4")
```

```python
#bar plot
categories=['A','B','C','D','E']
values=[5,7,3,8,6]

#create a bar plot
plt.bar(categories,values,color='blue')
```
